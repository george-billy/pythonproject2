```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('http://data.jakarta.go.id/dataset/8970f5d3-519d-4def-8cf4-47a71dad3baa/resource/75332e35-199d-45cf-a19f-985b16b46a77/download/Target-dan-Realisasi-Transmigran-Menurut-Daerah-Asal-dan-Jenisnya-2011.csv')
```

```python
iris.groupby('realisasi_kk').mean()
```

```python
iris.groupby('realisasi_orang').mean()
```

```python
plt.figure(figsize=(8,6))
plt.plot(iris['realisasi_kk'], label='KK')
plt.plot(iris['realisasi_orang'], label='Orang')
           
plt.xlabel('Data ke-')
plt.ylabel('Nilai')
plt.title('Data Realisasi Transmigran Menurut Jenisnya di Provinsi DKI Jakarta')
plt.legend()
plt.show()
```
